# Representación gráfica de probabilidades con Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Creando una gráfica usando matplotlib
x = range(0, 10)
y = [2 * value for value in x]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x,y)
plt.axis('equal')
plt.show()
```

```python
# Obtenemos la gráfica con los puntos a graficarse
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

Podemos 'subrayar' un intervalo entre dos valores de x por ejemplo.

```python
def is_in_interval(value, lower, upper):
    return lower <= value <= upper

plt.plot(x, y)
where = [is_in_interval(value, 2, 6) for value in x]
plt.fill_between(x, y, where=where)
plt.show()
```

```python
# Añadiendo títulos a la gráfica
plt.plot(x,y)
plt.title('Pirmera gráfica x vs. y')
plt.xlabel('Values between zero and ten')
plt.ylabel('Twice the values of x')
plt.show()
```
